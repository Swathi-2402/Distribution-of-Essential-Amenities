This code uses `geopandas` and `matplotlib` to plot the spatial distribution of amenities (hospitals, schools, supermarkets, companies) from GeoJSON data. Each amenity type is represented by different markers and colors. Conditional checks ensure column existence, and the plot includes a legend and labels for clarity.

```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Load GeoJSON data
file_url = "https://raw.githubusercontent.com/sushantkarn/DistributionofEssentialAmenities/main/export.geojson"
gdf = gpd.read_file(file_url)

# Plot the data using geopandas
fig, ax = plt.subplots(figsize=(10, 10))

# Check if 'amenity' column exists before plotting
if 'amenity' in gdf.columns:
    gdf[gdf['amenity'] == 'hospital'].plot(ax=ax, color='red', marker='H', markersize=50, label='Hospitals')
    gdf[gdf['amenity'] == 'school'].plot(ax=ax, color='green', marker='^', markersize=50, label='Schools')

# Check if 'shop' column exists before plotting
if 'shop' in gdf.columns:
    gdf[gdf['shop'] == 'supermarket'].plot(ax=ax, color='blue', marker='s', markersize=50, label='Supermarkets')

# Check if 'office' column exists before plotting
if 'office' in gdf.columns:
    gdf[gdf['office'] == 'company'].plot(ax=ax, color='purple', marker='o', markersize=50, label='Companies')

# Add legend and labels
ax.legend()
ax.set_title('Distribution of Amenities')
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')

# Add space between subplots
plt.subplots_adjust(bottom=0.1)

# Show the plot
plt.show()
```

This code uses `matplotlib` and `seaborn` to create a bar chart displaying the counts of essential amenities (schools, hospitals, grocery stores, companies) in Leeds. The data is organized in a Pandas DataFrame, and the chart is labeled with titles and axis labels for clarity. The resulting chart provides a quick visual summary of amenity distribution.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Sample data (replace this with your actual data)
data = {
    'Amenity': ['Schools', 'Hospitals', 'Grocery Stores', 'Companies'],
    'Count': [150, 30, 80, 100]
}

df = pd.DataFrame(data)

# Bar Chart
plt.figure(figsize=(10, 6))
sns.barplot(x='Amenity', y='Count', data=df)
plt.title('Distribution of Essential Amenities in Leeds')
plt.xlabel('Amenity')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

This code uses `matplotlib` to generate a pie chart illustrating the percentage distribution of essential amenities (schools, hospitals, grocery stores, companies) in Leeds based on the provided counts in the DataFrame, with labels and percentages displayed.

```python
# Pie Chart
plt.figure(figsize=(8, 8))
plt.pie(df['Count'], labels=df['Amenity'], autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Essential Amenities in Leeds')
plt.show()
```

This code creates a heatmap using `matplotlib` and `seaborn` to illustrate the distribution of essential amenities in Leeds. The DataFrame is reshaped, and the heatmap displays counts with color intensity, using the 'YlGnBu' colormap and annotations for clarity.

```python
# Heatmap
plt.figure(figsize=(10, 6))
heatmap_data = df.pivot(index=None, columns='Amenity', values='Count')
sns.heatmap(heatmap_data, cmap='YlGnBu', annot=True, fmt=".1f", linewidths=.5)  # Use ".1f" for one decimal place
plt.title('Heatmap of Essential Amenities in Leeds')
plt.show()
```

This code generates a stacked bar chart using `matplotlib`, depicting the counts of essential amenities in Leeds. The chart showcases the total counts for each amenity type, providing a clear visual representation.

```python
# Stacked Bar Chart
stacked_data = df.set_index('Amenity')
stacked_data.T.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Stacked Bar Chart of Essential Amenities in Leeds')
plt.xlabel('Amenity')
plt.ylabel('Count')
plt.show()
```

This code uses `matplotlib` and `seaborn` to generate an area chart illustrating the counts of essential amenities in Leeds. The chart combines a line plot with markers and a filled area beneath the line, providing a succinct overview of amenity count trends.

```python
# Area Chart
plt.figure(figsize=(10, 6))
sns.lineplot(x='Amenity', y='Count', data=df, marker='o')
plt.fill_between(df['Amenity'], df['Count'], color='skyblue', alpha=0.4)
plt.title('Area Chart of Essential Amenities in Leeds')
plt.xlabel('Amenity')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

This code uses Plotly Express to create a scatter chart, visualizing the counts of essential amenities (schools, hospitals, grocery stores, companies) in Leeds. Each amenity type is represented by markers, offering a clear overview of count distribution.
